```python
import matplotlib.pyplot as plt
```

```python
colors = ["red", "blue", "green", "yellow"]
```

```python
filter_dep = (df['dep'] == 750)
```

```python
fig, ax = plt.subplots(figsize=(16,8), nrows=1, ncols=1)

for (name, group), color in zip(df[filter_dep].groupby(['grav']), colors):
    
    group.plot(kind='scatter',
               x='long', 
               y='lat', 
               label=name, 
               color=color,
               ax=ax)
```
